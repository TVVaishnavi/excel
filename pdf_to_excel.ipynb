```python
import fitz
import pandas as pd

dp=fitz.open('1.pdf')
text=""
for page in dp:
    text+=page.get_text()


Taking_table=text.split()
Taking_Question=text.split("\n")


pfd_len_table=len(Taking_table) 
pfd_len_Question=len(Taking_Question)

def find_Index(name):#for Questons
    i=0
    while i<=pfd_len_Question:
        if Taking_Question[i]==name:
          return i
        i+=1

def find_index_table(name):#for tables 
    i=0
    while i<=pfd_len_table:
        if Taking_table[i]==name:
            return i
        i+=1

def find_Question(start,end):#find para
    paragraph=[]
    y=end-6
    x=start+5
    while x<y:
        paragraph.append(Taking_Question[x])
        x+=1
    return paragraph

def add_para():#prompt
    for i in cut_the_paragraph():
        print(i)

def find_para(start,end):#find explanation
    paragraph=""
    y=end-1
    x=start+3
    while x<y:
        paragraph+=(Taking_Question[x])
        x+=1
    return paragraph



Qdf="Assessment"
Domain="Domain"
skill="Skill"
ID="ID"
A="A."
B="B."
C="C."
D="D."
Correct="Answer:"


Question=find_Question(find_Index(skill),pfd_len_Question)
Question_len=len(Question)

I_ID_K=Taking_table[find_index_table(ID)+1]
I_Deff_K=Taking_table[find_index_table(Qdf)-1]
I_Domain_k=Taking_table[find_index_table(Domain)+1]+" "+Taking_table[find_index_table(Domain)+2]+" "+Taking_table[find_index_table(Domain)+3]
I_Skill_K=Taking_table[find_index_table(skill)+1]+" "+Taking_table[find_index_table(skill)+2]+" "+Taking_table[find_index_table(skill)+3]
I_Small_Question=Question[Question_len-1]
I_A_k=Taking_table[find_index_table(A)+1]
I_B_k=Taking_table[find_index_table(B)+1]
I_C_k=Taking_table[find_index_table(C)+1]
I_D_k=Taking_table[find_index_table(D)+1]
I_correct_k=Taking_table[find_index_table(Correct)+1]
para_start="Question ID {}".format(I_ID_K)
I_Explanation_k=find_para(find_Index(para_start),find_Index(Qdf))




def cut_the_paragraph():#split para
    paragraph=""
    x=0
    while x<Question_len-1:
        paragraph+=Question[x]
        x+=1
    return paragraph





data={'Question ID':[I_ID_K],'Domain':[I_Domain_k],'Skill':[I_Skill_K],'Question Difficulty':[I_Deff_K],'Question':[cut_the_paragraph()],'prompt':[I_Small_Question],'choice A':[I_A_k],'Choice B':[I_B_k],'Choice C':[I_C_k],'Choice D':[I_D_k],'Correct Answer':[I_correct_k],'Rationale/Explanation':[I_Explanation_k]}
df = pd.DataFrame(data)
excel_file_path=r'C:\Users\Vaishnavi\OneDrive\문서\output.xlsx'

df.to_excel(excel_file_path,index=False)
print(f"Data saved to{df.to_excel}")


```
